```python
import GetOldTweets3 as got
import requests
import json 
import pandas as pd
from time import sleep
import time
import tweepy
```

```python
hou_traffic_accounts = ['houstontranstar', 'us290nw_traffic', 'TxDOTHouston', 'TotalTrafficHOU', 'METROHouAlerts']
hou_public_safety_accounts = ['HoustonOEM', 'ReadyHarris', 'HCSOTexas', 'houstonpolice', 'HoustonFire']
hou_weather_accounts = ['NWSHouston', 'KHOUweather', 'abc13weather', 'KPRC2WEATHER']
```

```python
t0 = time.time()

hou_traffic_aug = []
for account in hou_traffic_accounts:
    tweetCriteria = got.manager.TweetCriteria().setUsername(account).setSince("2017-08-01").setUntil("2017-08-31")
    tweet = got.manager.TweetManager.getTweets(tweetCriteria) 
    for i in range(len(tweet)):
        tweet_dict = {}
        tweet_dict['username'] = tweet[i].username
        tweet_dict['date'] = tweet[i].date
        tweet_dict['text'] = tweet[i].text
        tweet_dict['hashtags'] = tweet[i].hashtags
        
        tweet_dict['type'] = 'official'
        hou_traffic_aug.append(tweet_dict)

print(f'Seconds to execute:',(time.time() - t0))
```
